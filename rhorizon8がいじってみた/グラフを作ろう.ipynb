```python
import pandas
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
csv = pandas.read_csv(".tsvとか糞食らえ.csv").values.tolist()
```

```python
usp = [0 for i in range(len(csv))]
for i in range(len(csv)):
    usp[i] = csv[i][14]
usp

fig2, bx = plt.subplots()
fig2.subplots_adjust(left=0, right=1, bottom=0.05, top=0.95)
bx.plot(range(len(usp)),usp, color="r")
bx.set_xlabel("id")
bx.set_ylabel("利用者数")
bx.set_title("大阪市における利用者予測結果")
fig2.set_figheight(5)
fig2.set_figwidth(7)
plt.show()
```

```python
hr = [[0 for i in range(3)] for j in range(24)]
hrp = [0 for i in range(24)]
for i in range(len(csv)):
    hr[csv[i][5]][0] += csv[i][14]
    hr[csv[i][5]][1] += 1

for i in range(len(hr)):
    hrp[csv[i][5]] = hr[csv[i][5]][0] / hr[csv[i][5]][1]


fig, ax = plt.subplots()
ax.plot(range(len(hrp)),hrp)
ax.set_xlabel("時間帯")
ax.set_ylabel("平均利用者数")
ax.set_title("時間帯別の平均利用者数")
fig.set_figheight(5)
fig.set_figwidth(7)
plt.show()
```
