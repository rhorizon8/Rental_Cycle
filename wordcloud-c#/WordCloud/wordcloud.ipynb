```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
f_path = '/usr/share/fonts/opentype/ipafont-gothic/ipag.ttf'
wakati=""

with open("./wakati.txt", mode="r") as f:
    wakati+=f.read()

wordcloud = WordCloud(background_color="white",
                      font_path=f_path,
                      width=700,
                      height=600).generate(wakati)
plt.figure(figsize=(12,10))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
