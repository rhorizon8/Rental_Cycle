```python
import pandas
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
csv = pandas.read_csv("tokyo.csv").values.tolist()
```

```python
usp = [0 for i in range(len(csv))]
for i in range(len(csv)):
    usp[i] = csv[i][14]
usp

fig2, bx = plt.subplots()
fig2.subplots_adjust(left=0, right=1, bottom=0.05, top=0.95)
bx.plot(range(len(usp)),usp, color="r")
bx.set_xlabel("id")
bx.set_ylabel("")
bx.set_title("")
fig2.set_figheight(5)
fig2.set_figwidth(7)
plt.show()
```

```python
hr_work = [[0 for i in range(2)] for j in range(24)]
hr_holy = [[0 for i in range(2)] for j in range(24)]
hrp_work = [0 for i in range(24)]
hrp_holy = [0 for i in range(24)]
for i in range(len(csv)):
    if int(csv[i][8]) == 0:
        hr_work[csv[i][5]][0] += csv[i][14]
        hr_work[csv[i][5]][1] += 1
    else :
        hr_holy[csv[i][5]][0] += csv[i][14]
        hr_holy[csv[i][5]][1] += 1

for i in range(len(hr_work)):
    hrp_work[csv[i][5]] = hr_work[csv[i][5]][0] / hr_work[csv[i][5]][1]
    hrp_holy[csv[i][5]] = hr_holy[csv[i][5]][0] / hr_holy[csv[i][5]][1]


fig, ax = plt.subplots()
ax.plot(range(len(hrp_work)),hrp_work,label="")
ax.plot(range(len(hrp_holy)),hrp_holy,label="")
ax.set_xlabel("")
ax.set_ylabel("")
ax.set_title("")
fig.set_figheight(5)
fig.set_figwidth(7)
ax.legend()
plt.show()
```
